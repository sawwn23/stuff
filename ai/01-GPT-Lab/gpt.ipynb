```python
with open ('input.txt', 'r', encoding='utf-8') as f:
  text = f.read()
```

```python
# get uniq chars on dataset text
chars = sorted(list(set(text)))
vocab_size = len(chars)
print(''.join(chars))
print(vocab_size)
```

# Before training need to prepare dataset into int since tensor is float ML understanable format 
# there are many toeknizers eg sentencepiece - google
# openai tiktoken
https://github.com/google/sentencepiece
https://github.com/openai/tiktoken

```python
# create a mapping of char to int (char index)
stoi = { ch:i for i,ch in enumerate(chars) }
itos = { i:ch for i,ch in enumerate(chars) }
encode = lambda s: [stoi[c] for c in s]
decode = lambda l: ''.join([itos[i] for i in l])

print(encode("hello world"))
print(decode(encode("hello world")))
```
